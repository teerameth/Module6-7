```python
from IPython.display import Image 
import cv2
import numpy as np
import math
from matplotlib import pyplot as plt
import PIL.Image
def imshow(img):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
img = cv2.imread("img/salt-and-pepper.jpg")
## LOW-pass filtering ##
kernel = np.ones((5,5), np.float32)/25
dst = cv2.filter2D(img, -1, kernel)
imshow(dst)
```

# Unsharp Masking / High-boost Filtering
![Unsharp Masking / High-boost Filtering](img/Task6.jpg)

```python
## Original signal ##
img = cv2.imread("img/Task6.jpg")
## Blurred signal ##
blurred = cv2.GaussianBlur(img, (7, 7), 0.0)
imshow(blurred)
```

```python
## Unsharp ##
unsharp = cv2.addWeighted(img, 2, blurred, -1, 0, img)
cv2.imwrite("img/Task6_unsharp.jpg", unsharp)
imshow(unsharp)
## High-boost ##
high_boost = cv2.addWeighted(img, 2.5, blurred, -1.5, 0, img)
cv2.imwrite("img/Task6_high_boost.jpg", high_boost)
imshow(high_boost)
```
