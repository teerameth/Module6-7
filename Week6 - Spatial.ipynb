```python
from IPython.display import Image 
import cv2
import numpy as np
import math
from matplotlib import pyplot as plt
import PIL.Image
def imshow(img):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
img = cv2.imread("img/salt-and-pepper.jpg")
## LOW-pass filtering ##
kernel = np.ones((5,5), np.float32)/25
dst = cv2.filter2D(img, -1, kernel)
imshow(dst)
```

```python
## Median Filtering ##
img = cv2.imread("img/salt-and-pepper.jpg")
imshow(img)
```
